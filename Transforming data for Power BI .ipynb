```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
"""
Data Fields
datetime - hourly date + timestamp  
season -  1 = spring, 2 = summer, 3 = fall, 4 = winter 
holiday - whether the day is considered a holiday
workingday - whether the day is neither a weekend nor holiday
weather - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
temp - temperature in Celsius
atemp - "feels like" temperature in Celsius
humidity - relative humidity
windspeed - wind speed
casual - number of non-registered user rentals initiated
registered - number of registered user rentals initiated
count - number of total rentals
"""
```

```python
def season(value):
    if value==1:
        return 'Spring'
    elif value==2:
        return 'Summer'
    elif value==3:
        return 'Fall'
    else:
        return 'Winter'       #Season has either of the 4 values so we use ELSE
```

```python
def weather(value):
    if value==1:
        return 'Clear'
    elif value==2:
        return 'Mist'
    elif value==3:
        return 'Sleet'  #Snow and Rain
    else:
        return 'Heavy Rain'
```

```python
data['weather']=data['weather'].apply(weather)
data['season']=data['season'].apply(season)
```

```python
data.head()
```

```python
data.info()
```

```python
def holiday(value):
    if value==0:
        return 'No'
    else:
        return 'Yes'
```

```python
def workingday(value):
    if value==0:
        return 'No'
    else:
        return 'Yes'
```

```python
data['holiday']=data['holiday'].apply(holiday)
data['workingday']=data['workingday'].apply(workingday)
```

```python
data.head()
```

```python
'2011-01-01 00:00:00'.split(' ')
```

```python
def date(value):
    return value.split(' ')[0]

def time(value):
    return value.split(' ')[1]
```

```python
data['date']=data['datetime'].apply(date)
data['time']=data['datetime'].apply(time)
```

```python
data.drop('datetime',axis=1,inplace=True)
```

```python
data
```

```python
data=data[['date','time','season','holiday','workingday','weather','temp','atemp','humidity','windspeed','casual','registered','count']]
```
